```python
# 1) Program to convert kilometres to miles
km = int(input("enter the kilometer"))

#converting into miles
if km<0:
  print("enter a positive number")
else:
  m = km * 0.621
  print(km,"km is equivalent to",m,"miles")
```

```python
# 2) Program to celcius to Fahrenheit 
c= int(input("enter the temperature in celcius"))
#conversion
f = (c*(9/5))+32
print(c,"C is equivalent to",f,'F')
```

```python
# 3) Program to display calender 
import calendar

y = int(input("enter the year"))  
m = int(input("enter the month"))  

# display the calendar
print(calendar.month(y, m))

```

```python
# 4) program to solve quadratic equation

print("a quadratic equation is of the form ax^2 + bx + c = 0 ")

import cmath
a = int(input("enter value for a "))
b = int(input("enter value for b "))
c = int(input("enter value for c "))

d = (b**2) - (4*a*c)

sol1 = (-b-cmath.sqrt(d))/(2*a)
sol2 = (-b+cmath.sqrt(d))/(2*a)

print('The two solutions are {0} and {1}'.format(sol1,sol2))
```

```python
# 5) Program to swap two variables without using temp variable
a = int(input("enter the first number"))
b = int(input("enter the second number"))
#before swaping
print("before swaping",'a=',a,'b=',b)
(a,b)=(b,a)
print("after swaping", 'a=',a,'b=',b)
```
